```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load dataset into dataframe for analysis

df = pd.read_csv("data/ames.csv", index_col=0)
df.head()
```

```python
# code to check for the shape of the dataframe
df.shape
```

```python
# checking datatypes for data in dataframe
df.info()
```

```python
# computing basic statistics of numerical columns
df.describe()
```

* Data Cleaning

```python
# checking for missing values in columns
df.isna().sum()
```

```python
# looking further into the missing values on the dataframe
df.isna().sample(5, random_state=1)
```

```python
# drop columns with more than 50% missing values
threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)
df.head()
```

```python
# filling in all numeric columns with their median values
num_cols = df.select_dtypes(include=['number']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# filling in object columns with 'None'
text_cols = df.select_dtypes(include=['object']).columns
df[text_cols] = df[text_cols].fillna('None')

# checking if the missing values have been filled in
print(df.isnull().values.any())
```

* Grouping 

```python
# grouping by neighborhood vs saleprice
grouped = df.groupby('Neighborhood')['SalePrice'].mean().sort_values(ascending=False)
print(grouped)
```

* Visualizations

```python
# grouping by yearbuilt for visualization
yearly_prices = df.groupby('YearBuilt')['SalePrice'].mean()

plt.figure(figsize=(12,6))
sns.lineplot(x=yearly_prices.index, y=yearly_prices.values)
plt.title('Average Sale Price by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Average Sale Price')
plt.grid(True)
plt.show()
```

* Neighborhood vs avg saleprices

```python
plt.figure(figsize=(14,7))
grouped.plot(kind='bar', color='skyblue')
plt.title('Average Sale Price per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average Sale Price')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

* Histogram for distribution of sales

```python
plt.figure(figsize=(10,6))
sns.histplot(df['SalePrice'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

* Scatter Plot for relationship between Living Area and Sale Price

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, hue='OverallQual', palette='viridis')
plt.title('Living Area vs Sale Price')
plt.xlabel('Above Grade (Ground) Living Area (sq ft)')
plt.ylabel('Sale Price')
plt.legend(title='Overall Quality')
plt.show()
```
